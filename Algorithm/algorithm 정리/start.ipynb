### 문제 해결 전략
1. 완벽한 문제 이해
> 하나하나 꼼꼼히 한 문장도 빼놓지 않고

2. 계획 / 설계 
> A4용지, 펜, 테블릿

3. 설계 한대로 구현 -> 디버깅

```python
N = int(input())
lst = [[0]*7 for _ in range(2)]

t1 = N
for i in range(4):
    lst[0][i] = t1
    t1 += 1

t2 = N
for j in range(6, 2, -1):
    lst[1][j] = t2
    t2 -= 1
    
    
print(lst[0])
print(lst[1])
```

### 복잡도 분석
1. 알고리즘의 효율
- 공간적 효율성과 시간적 효율성
    - 공간적 효율성 : 연산량 대비 얼마나 적은 메모리 공간(변수, 리스트 등)을 요하는가
    
    - 시간적 효율성 : 연산량 대비 얼마나 적은 시간을 요하는가 (반복(문)을 얼마나 많이 사용했는가)

    - 효율성을 뒤집어 표현하면 복잡도(Complexity). 복잡도가 높을수록 효율성은 저하

2. O(Big-Oh)-표기
- 복잡도의 점근적 상한을 나타냄

- 복잡도가 f(n) = 2n^2 - 7n + 3 -> O(n^2) >> 상수는 배제함

- 복잡도가 낮을수록 효율성이 좋다 -> O(n) > O(n^2)






- 로그 시간(O(logn)) / 로그 선형 시간(O(nlohn)) 중요
    > 오더 nlogn 이 효율성이 제일 좋다

- queue -> BFS
    - pop(0) -> 시간 복잡도 O(n)

    - deque 모듈로 popleft() -> O(1) >> 따라서 O(1)을 사용하는 팝 레프트를 사용하자!!

    - O(logn) -> 이진탐색

    - O(nlogn) -> sort ..
