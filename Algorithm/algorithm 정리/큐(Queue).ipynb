### 큐의 특성
> 스택과 마찬가지로 삽입과 삭제의 위치가 제한적인 자료구조

- 큐의 뒤에서는 삽입만 하고, 큐의 앞에서는 삭제만 이루어지는 구조


> 선입선출구조 (FIFO : First In First Out)

- 큐에 삽입한 순서대로 원소가 저장되어, 가장 먼저 삽입(first in)된 원소는 가장 먼저 삭제(first out) 된다.

- 후입선출과 같이 특징을 꼭 기억하자!!! 후입선출은 LIFO(last in first out)

> rear : 마지막 저장 위치

> front : 마지막으로 꺼내진 위치

### 선형큐
> 1차원 배열을 이용한 큐

- 큐의 크기 = 배열의 크기

- front : 마지막으로 꺼내진(삭제된) 인덱스

- rear : 저장된 마지막 원소의 인덱스

> 상태 표현

- 초기 상태 : front = rear = -1

- 공백 상태 : front == rear

- 포화 상태 : rear == n-1 (n: 배열의 크기, n-1: 배열의 마지막 인덱스)

##### 초기 공백 큐 생성
- 크기 n인 1차원 배열 생성

- front와 rear를 -1로 초기화

````python
N = 10          # 큐 생성
q = [0] * 10
front = rear = -1

rear += 1       # enqueue(10)
q[rear] = 10
rear += 1       # enqueue(20)
q[rear] = 20
rear += 1       # enqueue(30)
q[rear] = 30
while front != rear:   # 큐가 비어있지 않으면
    front += 1          # dequeue()
    print(q[front])
````

### 원형큐
> 초기 공백 상태

- front = rear = 0

> Index의 순환

- front와 rear의 위치가 배열의 마지막 인덱스인 n-1를 가리킨 후, 그 다음에는 논리적 순환을 이루어 배열의 처음 인덱스인 0으로 이동해야 함

- 이를 위해 나머지 연산자 mod를 사용함









> front == (rear+1)%N -> 큐가 꽉 찬 상태인지 확인하는 방법

---

##### 원형큐의 공백상태 및 포화상태 검사: isEmpty(), isFull()
- 공백상태 : front == rear

- 포화상태 : 삽입할 rear의 다음 위치 == 현재 front
    - (rear + 1)mod n == front

### 연결큐


> linked list는 스택, 큐, 트리 구조등 여러 구조에 사용 가능

##### 연결 큐의 연산 과정

> null : 어느 위치도 가리키고 있지 않음







````python
from collections import deque

q = deque()
q.append(1)
q.append(2)
print(q.popleft())
print(q.popleft())
# ---
q = []
for i in range(10000):
    q.append(i)
print('append')
while q:
    q.pop(0)  # 다 꺼내라!
    print('end')
#---
dq = deque()
for i in range(10000):
    q.append(i)
print('append')
while dq:
    dq.popleft()
print('end')
````

### 우선순위 큐

> 우선순위 큐의 특성
    
- 우선순위를 가진 항목들을 저장하는 큐

- FIFO 순서가 아니라 우선순위가 높은 순서대로 먼저 나가게 됨


> 우선순위 큐의 적용 분야

- 시뮬레이션 시스템

- 네트워크 트래픽 제어

- 운영체제의 테스크 스케줄링

> 우선순위 큐의 구현

- 배열을 이용한 우선순위 큐

- 리스트를 이용한 우선순위 큐

> 우선순위 큐의 기본 연산

- 삽입 : enQueue

- 삭제 : deQueue



#### 배열을 이용한 우선순위 큐
> 배열을 이용하여 우선순위 큐 구현

- 배열을 이용하여 자료 저장

- 원소를 삽입하는 과정에서 우선순위를 비교하여 적절한 위치에서 삽입하는 구조

- 가장 앞에 최고 우선순위의 원소가 위치하게 됨

> 문제점

- 배열을 사용하므로, 삽입이나 삭제 연산이 일어날 때 원소의 재배치가 발생함

- 이에 소요되는 시간이나 메모리 낭비가 큼

### 버퍼 (Buffer)
- 데이터를 한곳에서 다른 한 곳으로 전송하는 동안 일시적으로 그 데이터를 보관하는 메모리의 영역

- 버퍼링 : 버퍼를 활용하는 방식 또는 버퍼를 채우는 동작을 의미

- 버퍼의 자료구조
    - 버퍼는 일반적으로 입출력 및 네트워크와 관련된 기능에서 이용 됨

    - 순서대로 입력/출력/전달 되어야 하므로 FIFO 방식의 자료구조인 큐가 활용 됨
