##### 백트래킹과 깊이우선탐색과의 차이


##### 백트래킹을 이용한 알고리즘 절차
1. 상태 공간 트리의 깊이 우선 검색을 실시

2. 각 노드가 유망한지를 점검

3. 만일 그 노드가 유망하지 않으면, 그 노드의 부모 노드로 돌아가서 검색을 계속

##### 깊이 우선 검색 vs. 백트래킹
- 순수한 깊이 우선 검색 = 155 노드

- 백트래킹 = 27 노드

```python
# 부분집합
def f(i, k):
    if i == k:
        for j in range(k):
            if bit[j]:
                print(arr[j], end = ' ')
        print()

    else:
        # for j in range(2):
        #     bit[1] = j
        #     f(i+1, k)
        bit[i] = 1
        f(i+1, k)
        bit[i] = 0
        f(i+1, k)



N = 4
arr = [1, 2, 3, 4]
bit = [0] * N    # bit[i] : arr[i]가 부분집합에 포함되었는지 나타내는 배열
f(0, N)     # bit[i]에 1또는 0을 채우고, N개의 원소가 결정되면 부분집합을 출력
```
