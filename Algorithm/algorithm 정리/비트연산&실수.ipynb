# 비트 연산
- 1 bit : 0과 1을 표현하는 정보의 단위

- 1 Byte : 8 bit를 묶어 1 Byte 라고 함

> 컴퓨터의 CPU는 0과 1로 동작. 내부적으로 비트연산을 사용하여 덧셈, 뺄셈, 곱셈 등을 계산

### AND 와 OR
- & : num1 & num2
- | : num1 | num2





```python
print(0b11011110 & 0b11011)
print(0x4A3 | 25)
```

### XOR 연산자
> ^ : XOR(엑스오어) 연산자, OR 처럼 동작되는데 둘 다 1인 경우는 0이다.

- 같으면 0 / 다르면 1







```python
# 암호화 프로그램 제작하기
"""
암호화를 해주거나(인코딩) 암호를 해제해주는(디코딩) 프로그램(함수로) 제작
"""
def encode_decode(num):
    return num ^ KEY


print(encode_decode(1000))
print(encode_decode(4))
num = int(input())
KEY = 1004
```

### Left와 Right Shift 연산자
- Left Shift << : 특정 수 만큼 비트를 왼쪽으로 밀어낸다. (우측에 0이 생성)

- Right Shift >> : 특정 수 만큼 비트를 오른쪽으로 밀어낸다. (우측 비트들이 제거)



```python
# Left Shift (<<)를 이용한 프로그래밍
# 반복문 사용
t = 1
for i in range(5):
    print(bin(t), t)
    t = t << 1

# 출력 결과는 2의 제곱수 이다
```

> 비트연산 응용 1
- 1 << N  -> 2^n 값을 가진다


> 비트연산 응용 2
- i & (1<<n)



> i의 n번 비트가 1인지 아닌지를 확인할 수 있다.

### NOT 연산자
- (~) NOT 연산자 : 모든 비트를 반전시킨다

- 만약 8 bit 일때 ~ (0001 1111) 이라면 값은 1110 0000이 된다

- 파이썬에서 ~4를 수행하면 -5가 출력됨





```python
# swea 10726 이진수 표현
"""
정수 N, M이 주어질 때, M의 이진수 표현의 마지막 N 비트가 모두 1로 켜져있는지
아닌지를 판별하여 출력, 모두 켜있다면 ON 출력, 아니면 OFF 출력
"""
# 내가 짠 코드
# N, M = map(int, input().split())
# for i in range(N):
#     if M[-1 : -N] == 1:
#         print('ON')
#     else:
#         print('NO')
"""
그저 슬라이싱을 통해 푸려고 접근
비트 연산자로 했어야함
"""

M = 31
N = 5
def Test():
    tar = M
    for i in range(N):
        if tar & 0x1 == 0:
            return False
        
        tar = tar >> 1
    return True
print(Test())
```
